```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
conn=sqlite3.connect("database.sqlite")
c=conn.cursor()
```

```python
#full description of the tables inside the database that we have
tables = pd.read_sql("""SELECT *
                        FROM sqlite_master
                        WHERE type='table';""", conn)
tables
```

```python
#get the schema of the table
pd.read_sql("PRAGMA table_info('Reviews');",conn)
```

```python
pd.read_sql("SELECT * FROM Reviews;",conn)
```

```python
df_table= pd.read_sql("""SELECT * FROM Reviews;""", conn)
```

```python
#to convert Time values to readble time value:
df_table['Date_Time']= pd.to_datetime(df_table['Time'], unit='s')
```

```python
df_table
```

```python
#drop table if exists:
c.execute("DROP TABLE IF EXISTS table_reviews;")
```

```python
#convert data frame (pandas) to database table (sqlite):
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_sql.html
df_table.to_sql('table_reviews', conn, if_exists='fail')
```

```python
pd.read_sql("""SELECT * FROM table_reviews;""",conn)
```

```python
#get the positive reviews based on the scores (score>3 is positive):
pd.read_sql("""SELECT ProductId, UserId, Score, Summary FROM table_reviews 
                WHERE Score>=3 ORDER BY SCORE DESC, Summary ;""",conn)
```

```python
pd.read_sql("""SELECT Score, COUNT(Score) as scores FROM table_reviews
                WHERE Score>=3 GROUP BY score;""",conn)
```

```python
#get the positive reviews based on the review and score:
pd.read_sql("""SELECT ProductId, UserId, Score, Summary FROM table_reviews 
                WHERE Score=5 AND Summary LIKE "%good%" 
                OR Score=5 AND Summary LIKE "%perfect%" 
                OR Score=5 AND Summary LIKE "%delight%"
                OR Score=5 AND Summary LIKE "%nice%"                
                ORDER BY Summary ;""",conn)
```

```python
#find products those get 5 for the score and purchased more than 1 time:
pd.read_sql("""SELECT ProductId, Score, COUNT(*) FROM table_reviews
                WHERE Score=5 GROUP BY ProductId HAVING COUNT(*)>90
                ORDER BY ProductId DESC;""",conn)
```

```python
#sort product those have with score =5 by the number of purchase:
pd.read_sql("""SELECT ProductId, Score, COUNT(*) FROM table_reviews
                WHERE Score=5 GROUP BY ProductId
                ORDER BY COUNT(*) DESC;""",conn)
```

```python
#get the information of users those purchased the product that has score=5 
#and more than the other are purchased:
pd.read_sql("""SELECT UserId, Score, Summary FROM table_reviews
                WHERE ProductId="B007JFMH8M"
                ORDER BY UserId DESC;""",conn)
```

```python
#get the users whose purchased more than 10 times and give score 5 to their purchases:
pd.read_sql("""SELECT UserId, Score, COUNT(*) FROM table_reviews
                WHERE Score=5 GROUP BY UserId HAVING COUNT(*)>10
                ORDER BY COUNT(*) DESC;""",conn)
```
